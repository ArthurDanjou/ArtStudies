#### Dataframes

Le dataframe est l’**extension en 2 dimensions des series**. Il peut être vu comme un empilement de series dont les index sont partagés (et donc intrinsèquement alignés), ou comme dans un tableur où les index sont les numéros de lignes et les noms des series les noms des colonnes.

Un dataframe se comporte comme un dictionnaire dont les cléss sont les noms des colonnes et les valeurs sont des series.
On peut créer un dataframe à partir de dictionnaires, de liste de listes ou de liste de series, d’arrays numpy, de fichier Excel ou CSV, depuis des bases de données, …

### La bibliothèque NumPy

**La bibliothèque NumPy permet d’effectuer des calculs numériques avec Python. Elle introduit une gestion facilitée des tableaux de nombres.**

Pour info : Pandas ou Numpy ? Dans le cas de Pandas, ce ne sont pas tout-à-fait des matrices mais des tables de données : la même table peut contenir différents types de données, des nombres, des chaînes de caractères, des booléens, des dates… Les données se manipulent comme des bases de données et les Dataframe de pandas sont optimisées pour ça. Chaque colonne porte un nom. Numpy propose des matrices à une ou plusieurs dimensions optimisées pour le calcul matriciel. Ces matrices ne contiennent que des nombres et tous du même type.

**--> Pour faire des calculs matriciels, on choisit numpy. Pour faire de la gestion de base de données telle qu'une jointure de table, on choisit pandas.**

**Exercice :**
Ecrire un programme qui sépare les variables selon leur type (variables quantitatives, catégorielles et binaires). Pensez à identifier les variables avec des valeurs manquantes

```python
#Ecrivez votre code ici
quantitatives = []
categorielles = []
binaires = []

for col in data_set.columns:
    if data_set[col].dtype in ['int64', 'float64']:
        if len(data_set[col].dropna().unique()) == 2:
            binaires.append(col)
        else:
            quantitatives.append(col)
    else:
        if len(data_set[col].dropna().unique()) == 2:
            binaires.append(col)
        else:
            categorielles.append(col)

print("Variables quantitatives :", quantitatives)
print("\nVariables catégorielles :", categorielles)
print("\nVariables binaires :", binaires)
```

```python
#Ecrivez votre code ici
variables_avec_na = []

for col in data_set.columns:
    if data_set[col].isnull().any():
        variables_avec_na.append(col)

print("\nVariables avec valeurs manquantes :", variables_avec_na)
```

## 6. Analyse de la sinistralité 

Notre objectif final est de construire un modèle tarifaire. Il convient donc de construire 2 variables qui caractérisent la sinistralité :
* La fréquence de sinistres (NB / EXPO) 
* Le coût moyen (CHARGE / NB)

```python
#Calcul de la fréquence
data_retraitee["FREQ"] = data_retraitee["NB"] / data_retraitee["EXPO"]
data_retraitee["FREQ"] = data_retraitee["FREQ"].fillna(0)

#Calcul du coût moyen
data_retraitee["CM"] = data_retraitee["CHARGE"] / data_retraitee["NB"]
data_retraitee["CM"] = data_retraitee["CM"].fillna(0)
```

Les modèles tarifaires essayent d'établir un lien entre les variables explicatives et la sinistralité. C'est le travail d'un actuaire comprendre et avoir un regard critique sur les relations trouvées.  
Pour aider l'actuaire dans ce travail, la visualisation des données (FREQ, CM) en fonction des variables explicatives peut être utile.  
  
Dans ce qui suit, nous allons faire cet exercice que sur certaines variables explicatives. Cet exercice peut bien sûr être étendu aux autres variables explicatives.  
**Remarque :** 
*Attention au sens de l'analyse que vous menez. Si on s'intéresse par exemple au coût moyen de la garantie RC, ce dernier n'a pas de dépendance géographique (toute chose égale par ailleurs une victime sera indemnisée à la même hauteur à Lille ou à Marseille).*

 ### Fréquence en fonction de l'âge du conducteur

```python
#Ecrivez votre code ici
freq_by_age = data_retraitee.groupby('AGE_ASSURE_PRINCIPAL').agg({
    'NB': 'sum',
    'EXPO': 'sum'
}).reset_index()

freq_by_age['FREQ'] = freq_by_age['NB'] / freq_by_age['EXPO']
freq_by_age['FREQ'] = freq_by_age['FREQ'].fillna(0)

fig = px.line(freq_by_age, x='AGE_ASSURE_PRINCIPAL', y='FREQ',
              title='Fréquence de sinistres en fonction de l\'âge du conducteur')
fig.update_xaxes(title='Âge du conducteur principal')
fig.update_yaxes(title='Fréquence')
fig.show()
```

### Fréquence en fonction du genre

```python
#Ecrivez votre code ici
freq_by_age = data_retraitee.groupby('GENRE').agg({
    'NB': 'sum',
    'EXPO': 'sum'
}).reset_index()

freq_by_age['FREQ'] = freq_by_age['NB'] / freq_by_age['EXPO']
freq_by_age['FREQ'] = freq_by_age['FREQ'].fillna(0)

fig = px.line(freq_by_age, x='GENRE', y='FREQ',
              title='Fréquence de sinistres en fonction du genre du conducteur')
fig.update_xaxes(title='Genre du conducteur principal')
fig.update_yaxes(title='Fréquence')
fig.show()
```
