```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
lam = 1
nb_repl = 100000
sample = np.random.exponential(lam, nb_repl)
intervalle = np.linspace(0, 5, 100)
plt.hist(sample, bins=intervalle, density=True, label='Echantillon')

densite = stats.expon().pdf
plt.plot(intervalle, densite(intervalle), label='Fonction densit√©')
plt.legend()
```

```python
np_repl = 10000
sampleX, sampleY = 2*np.random.rand(np_repl)-1, 2*np.random.rand(np_repl)-1

intervalle = np.linspace(-2, 2, 100)
plt.hist(sampleX + sampleY, bins=intervalle, density=True, label='Echantillon')

densite = stats.uniform(-1, 2).pdf
plt.plot(intervalle, densite(intervalle), label='Fonction densit√©')

plt.plot(intervalle, 1/4 * np.maximum(2 - np.abs(intervalle), 0), label='Fonction')
plt.legend()
```

```python
lam = 1
nb_repl = 10000
sampleX, sampleY, sampleZ = np.random.exponential(lam, nb_repl), np.random.exponential(lam, nb_repl), np.random.exponential(lam, nb_repl)
intervalle = np.linspace(-5, 5, 100)
plt.hist(sampleX - sampleY/2, bins=intervalle, density=True, alpha=.7, label='Echantillon de X - Y/2')
plt.hist(sampleZ/2, bins=intervalle, density=True, alpha=.7, label='Echantillon de Z')
plt.legend()
```

```python
p = 1/2
nb_repl = 1000

plt.figure(figsize=(18, 5))
for i, k in enumerate([10, 100, 1000]):
    plt.subplot(1, 3, i+1)
    sample = np.random.binomial(k, p, nb_repl)
    intervalle = np.linspace(np.min(sample), np.max(sample), 100)
    plt.hist(sample, bins=intervalle, density=True, label=f'Echantillon de X pour n={k}')
    plt.legend()
```

```python
def sample_uniforme(N):
    return 2*np.random.rand(N) - 1
```

```python
nb_repl = 100_000
nb_lgn = 1000
liste_Sn = []

for _ in range(nb_lgn):
    liste_Sn.append(np.mean(sample_uniforme(nb_repl)))
    
nb_bins = 100
intervalles = np.linspace(np.min(liste_Sn), np.max(liste_Sn), nb_bins)
plt.hist(liste_Sn, density=True, bins=intervalles)
plt.show()

plt.figure()
nb_bins = 100
intervalles = np.linspace(-1, 1, nb_bins)
plt.hist(liste_Sn, density=True, bins=intervalles)
plt.show()
```
